Θα υλοποιήσουμε μια συνάρτηση για το κάθε ερώτημα

Για το __(α)__ φτιάχνω την def statistics(X,y) η οποία υπολογίζει το μέσο και τη διασπορά του κάθε χαρακτηριστικού για το κάθε ψηφίο.

Ο πίνακας για τις μέσες τιμές και τις διασπορές θα ειναι σε θέμα μεγέθους ίδιος , (10,256) , 10 γραμμές λόγω των 10 ψηφίων και 256 γραμμές λόγω των 256 χαρακτηριστικών

Η λογική της συνάρτησης είναι να γίνεται μια επανάληψη όσος και ο αριθμός των label , δηλαδή 10 (0-9) και να βρίσκω σε κάθε επανάληψη με τη χρήση mask στον πίνακα x_train τον αριθμό που είναι ίδιος με τον iterator της λούπας. Για παράδειγμα, αν είμαι στην πρώτη επανάληψη i=0 λόγω της εντολής:
                        __train[condition]__ όπου __condition=[y == i]__ θα πάρω τα στοιχεία 0 του train κ.ο.κ.
   

```python
def statistics(X,y):
    variance = np.zeros((10,256))
    mean = np.zeros((10,256))
    for i in range(10):
        condition = [y == i]
        digit = X[condition]
        mean[i] = digit.mean(axis=0)
        variance[i]=digit.var(axis=0)
    return mean,variance
```

# ΒΗΜΑ 12

- Παρακάτω φαίνεται ο  υλοποιημένος κώδικας σε scikit-learn estimator.
- Στη συνάρτηση init , θέτουμε το self.X_mean_ = None.
- To trailing underscore δεν είναι τίποτα παραπάνω από το να καταλαβαίνουμε ότι οι μεταβλητές αυτές αναφέρονται στο scope του class.
- Στη μέθοδο fit γίνεται το training του estimator μας, καθώς εκέι μαθαίνει τον πίνακα self.X_mean_ πάνω στον οποίο θα βασίζονται οι ταξινομήσεις μετά.
- Ο αλγόριθμος του fit βασίζεται στη λογική του βήματος 9 που υπολόγισα για κάθε ψηφίο το mean value του. Δηλαδή πάλι αρχικοποιώ το self.X_mean_ με 0 και μέγεθος (labels,X.shape[1]) όπου X.shape[1] είναι ο αριθμός στηλών του train dataset και το labels είναι ο αριθμός των κλάσεων που βρίσκει από το σημείο του κώδικα labels = len(set(y.tolist())). 
- Στη συνέχεια, τρέχω μια λούπα για κάθε class και με τη χρήση mask στο πινακα X κρατάω το ψηφίο για το οποίο θέλω να υπολογίσω τις μέσες τιμές του.
- Ο αλγόριθμος της συνάρτησης predict είναι αντίστοιχος του βήματος (11α) .
- Ενώ της συνάρτησης scorer είναι αντίστοιχος του βήματος (11β).

```python
from sklearn.base import BaseEstimator, ClassifierMixin
import numpy as np
class EuclideanClassifier(BaseEstimator, ClassifierMixin):  
    """Classify samples based on the distance from the mean feature value"""

    def __init__(self):
        self.X_mean_ = None
        #self.label_number_ = label_number


    def fit(self, X, y):
        """
        This should fit classifier. All the "work" should be done here.
        
        Calculates self.X_mean_ based on the mean 
        feature values in X for each class.
        
        self.X_mean_ becomes a numpy.ndarray of shape 
        (n_classes, n_features)
        
        fit always returns self.
        """
        labels = len(set(y.tolist()))
        self.X_mean_ = np.zeros((labels,X.shape[1]))
        for i in range(labels):
            condition = [y == i]
            digit = X[condition]
            self.X_mean_[i] = digit.mean(axis=0)       

        return self


    def predict(self, X):
        """
        Make predictions for X based on the
        euclidean distance from self.X_mean_
        """
        predictions=[]
        for i in range(X.shape[0]):
            dist = np.sqrt(np.sum((self.X_mean_-X[i])**2,axis=1))
            predictions.append(np.argmin(dist))
        
        return np.asarray(predictions)
    
    def scorer(self, X, y):
        """
        Return accuracy score on the predictions
        for X based on ground truth y
        """
        acc_arr = (X == y)
        return (np.sum(acc_arr) / X.shape[0]) * 100

```

```python
model = EuclideanClassifier()

model.fit(x_train,y_train)

predictions_from_myest = model.predict(x_test)

score_from_myest = model.scorer(predictions_from_myest,y_test)

print(score_from_myest)
```

```python
def plot_clf(clf, X, y, labels):
    fig, ax = plt.subplots()
    # title for the plots
    title = ('Decision surface of Classifier')
    # Set-up grid for plotting.
    X0, X1 = X[:, 0], X[:, 1]
    
    x_min, x_max = X0.min() - 1, X0.max() + 1
    y_min, y_max = X1.min() - 1, X1.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, .05),
                         np.arange(y_min, y_max, .05))
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    
    zeros = ax.scatter(
        X0[y == 0], X1[y == 0],
        c='blue', label=labels[0],
        s=60, alpha=0.9, edgecolors='k')
    ones = ax.scatter(
        X0[y == 1], X1[y == 1],
        c='red', label=labels[1], 
        s=60, alpha=0.9, edgecolors='k')
    twos = ax.scatter(
        X0[y == 2], X1[y == 2],
        c='yellow', label=labels[2], 
        s=60, alpha=0.9, edgecolors='k')
    threes = ax.scatter(
        X0[y == 3], X1[y == 3],
        c='pink', label=labels[3], 
        s=60, alpha=0.9, edgecolors='k')
    fours = ax.scatter(
        X0[y == 4], X1[y == 4],
        c='olive', label=labels[4], 
        s=60, alpha=0.9, edgecolors='k')
    fives = ax.scatter(
        X0[y == 5], X1[y == 5],
        c='yellow', label=labels[5], 
        s=60, alpha=0.9, edgecolors='k')
    sixs = ax.scatter(
        X0[y == 6], X1[y == 6],
        c='purple', label=labels[6], 
        s=60, alpha=0.9, edgecolors='k')
    sevens = ax.scatter(
        X0[y == 7], X1[y == 7],
        c='green', label=labels[7], 
        s=60, alpha=0.9, edgecolors='k')
    eights = ax.scatter(
        X0[y == 8], X1[y == 8],
        c='orange', label=labels[8], 
        s=60, alpha=0.9, edgecolors='k')
    nines = ax.scatter(
        X0[y == 9], X1[y == 9],
        c='brown', label=labels[9], 
        s=60, alpha=0.9, edgecolors='k')
    
    ax.set_ylabel(labels[1])
    ax.set_xlabel(labels[0])
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)
    ax.legend()
    plt.show()
```

```python
def plot_learning_curve(train_scores, test_scores, train_sizes, ylim=(0, 1)):
    plt.figure(5, figsize=(4,4))
    plt.title("Learning Curve")
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

```python
from sklearn.model_selection import learning_curve
train_sizes, train_scores, test_scores = learning_curve(EuclideanClassifier(), 
                                                        x_full, y_full, cv=5,
                                                        train_sizes=np.linspace(0.15, 1.0, 5))
```
